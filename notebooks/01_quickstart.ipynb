## 1. One-Line Forecasting

Predict the next 12 months of airline passengers. Vectrix automatically selects the best model from 30+ options.

```python
from vectrix import forecast, loadSample

df = loadSample("airline")
print(f"Data: {len(df)} rows, columns: {list(df.columns)}")
df.tail()
```

```python
result = forecast(df, date="date", value="passengers", steps=12)
print(result.summary())
```

```python
# Forecast values with confidence intervals
result.table()
```

```python
# Accuracy metrics
print(f"Best model: {result.model}")
print(f"MAPE: {result.mape:.2f}%")
print(f"RMSE: {result.rmse:.2f}")
print(f"MAE:  {result.mae:.2f}")
```

## 2. Time Series Analysis & DNA Profiling

Automatically detect trend, seasonality, anomalies, and 65+ features that form a unique DNA fingerprint.

```python
from vectrix import analyze

analysis = analyze(df, date="date", value="passengers")
print(analysis.summary())
```

```python
# DNA fingerprint — unique characteristics of this time series
dna = analysis.dna
print(f"Trend strength:      {dna.trendStrength:.3f}")
print(f"Seasonal strength:   {dna.seasonalStrength:.3f}")
print(f"Difficulty:          {dna.difficulty}")
print(f"Recommended model:   {dna.recommendedModels[0]}")
```

```python
# Anomalies detected
print(f"Anomalies found: {len(analysis.anomalies)}")
if len(analysis.anomalies) > 0:
    for a in analysis.anomalies[:5]:
        print(f"  Index {a['index']}: value={a['value']:.1f}, zscore={a['zscore']:.2f}")
```

## 3. R-Style Regression

Run regression with R-style formulas. Full diagnostics included.

```python
import pandas as pd
import numpy as np
from vectrix import regress

np.random.seed(42)
n = 200
reg_df = pd.DataFrame({
    "ads": np.random.uniform(100, 500, n),
    "price": np.random.uniform(10, 50, n),
    "season": np.tile(["spring", "summer", "fall", "winter"], n // 4),
})
reg_df["sales"] = 50 + 0.3 * reg_df["ads"] - 2.0 * reg_df["price"] + np.random.normal(0, 10, n)

reg = regress(data=reg_df, formula="sales ~ ads + price")
print(reg.summary())
```

```python
# Key statistics
print(f"R-squared:     {reg.r_squared:.4f}")
print(f"Adj R-squared: {reg.adj_r_squared:.4f}")
print(f"F-statistic:   {reg.f_stat:.2f}")
print(f"Durbin-Watson: {reg.durbin_watson:.4f}")
```

## 4. Compare Models Side by Side

Compare all available models on the same dataset.

```python
from vectrix import compare

comp = compare(df, date="date", value="passengers", steps=12)
print(comp.summary())
```

```python
# Comparison table: sMAPE, MAPE, RMSE, MAE for every model
comp.table()
```
